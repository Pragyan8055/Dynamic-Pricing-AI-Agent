```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
```

```python
adj_matrix = pd.read_csv('C:/Users/pragy/Downloads/Analytics_Summer_projects_2026/EV_dynamic_pricing_socbiz/UrbanEV_ SZ_districts/adj.csv', index_col=0)
dist = pd.read_csv('C:/Users/pragy/Downloads/Analytics_Summer_projects_2026/EV_dynamic_pricing_socbiz/UrbanEV_ SZ_districts/distance.csv', index_col=0)
nodes = adj_matrix.index.tolist()
adj_list = adj_matrix.values.tolist()
```

```python
adj_list = [adj_matrix.loc[node, :].values.flatten().tolist() for node in nodes]
G = nx.Graph()
for i, node in enumerate(nodes):
    for j, neighbor in enumerate(adj_list[i]):
        if neighbor == 1:
            G.add_edge(node, nodes[j], weight=dist.iloc[i, j+1])
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

```python
adj_m = np.array(adj_list)
G = nx.from_numpy_array(adj_m)

# 3. Plot the network
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

```python
degree_per_node = [adj_matrix[str(node)].mean() for node in nodes]

plt.figure(figsize=(10, 6))
sns.barplot(x=nodes, y=degree_per_node, palette='viridis')
plt.title('Average Degree per Node')
plt.xlabel('Node')
plt.ylabel('Average Degree')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# distribution of degree_per_node (kde)
plt.figure(figsize=(10, 6))
sns.kdeplot(degree_per_node, shade=True, color='blue')
plt.title('Distribution of Average Degree per Node')
plt.xlabel('Average Degree')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
G = nx.from_pandas_adjacency(adj_matrix)

# 3. Plot the network
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='skyblue', node_size=500, edge_color='gray', font_size=10)
plt.title("Network Graph from Adjacency Matrix")
plt.show()
```

```python
G = nx.from_pandas_edgelist(map_df, "source", "target", create_using=nx.Graph())

G.number_of_nodes(), G.number_of_edges()
```

```python
network = G
fig, ax = plt.subplots(figsize=(15, 9))
ax.axis("off")
plot_options = {"node_size": 10, "with_labels": False, "width": 0.15}
nx.draw_networkx(G, pos=nx.random_layout(G), ax=ax, **plot_options)
```

```python
degree_centrality = nx.centrality.degree_centrality(network)

plt.figure(figsize=(6, 3))
plt.hist(degree_centrality.values(), bins=25)
#plt.xticks(ticks=[0, 0.025, 0.05, 0.1, 0.15, 0.2])  # set the x axis ticks
plt.title("Degree Centrality Histogram ", fontdict={"size": 16}, loc="center")
plt.xlabel("Degree Centrality", fontdict={"size": 12})
plt.ylabel("Counts", fontdict={"size": 12})
plt.show()
```

```python
# graph visualization with node size proportional to degree centrality

node_size = [ v * 1000 for v in degree_centrality.values() ]  

pos = nx.spring_layout(network, iterations=1, seed=0)

plt.figure(figsize=(15, 8))
nx.draw_networkx(network, 
                 pos=pos, 
                 node_size=node_size, 
                 with_labels=False, width=0.15)
plt.axis("off")
plt.show()
```
